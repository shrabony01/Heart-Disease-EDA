```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('heart.csv')
display(df.head())
```

```python
# 1. Inspect the Data
print("Dataset Shape:", df.shape)
print("\nMissing Values:\n", df.isnull().sum())

# 2. Graph 1: The Target Distribution
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='HeartDisease', palette='Set2')
plt.title('Distribution of Heart Disease (Target Variable)')
plt.xlabel('Heart Disease (0 = No, 1 = Yes)')
plt.ylabel('Patient Count')
plt.show()
```

```python
# Graph 2: Continuous vs. Categorical (Age by Heart Disease)
plt.figure(figsize=(8, 5))
# Added 'hue' and 'legend=False' to fix that warning!
sns.boxplot(data=df, x='HeartDisease', y='Age', hue='HeartDisease', palette='Set1', legend=False)
plt.title('Age Distribution by Heart Disease Status')
plt.xlabel('Heart Disease (0 = No, 1 = Yes)')
plt.ylabel('Age')
plt.show()

# Graph 3: Feature Interaction (Chest Pain Type vs. Target)
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='ChestPainType', hue='HeartDisease', palette='viridis')
plt.title('Heart Disease Frequency by Chest Pain Type')
plt.xlabel('Chest Pain Type')
plt.ylabel('Patient Count')
plt.legend(title='Heart Disease', labels=['No', 'Yes'])
plt.show()
```
